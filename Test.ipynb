```python
from pyspark import SparkContext
```

```python
from pyspark import SparkConf
```

```python
from pyspark.sql import SQLContext
```

```python
conf =  SparkConf().setMaster("local[*]").setAppName("ReadCsv")
```

```python
sc = SparkContext(conf = conf)
```

```python
sq = SQLContext(sc)
```

```python
df = sq.read.csv("HR_Data.csv", header=True)
```

```python
print(type(df))
```

```python
df.printSchema
```

```python
df.show()
```

```python
df.filter(df["last_evaluation"] > .95).show()
```

```python
df1 = df.select("satisfaction_level","Work_accident", "sales")
```

```python
df1.printSchema()
```

```python
df1.show(30)
```

```python
tabList = df1.collect()
```

```python
print(type(tabList))
```

```python
for item in tabList:
    print(item["sales"])
```
