```python
import matplotlib.pyplot as plt
```

```python
year = [2015,2016,2017,2018,2019]
```

```python
sales = [130,120,170,180,200]
```

```python
profit = [30,40,50,60,20]
```

```python
plt.plot(year,sales, color='red')
plt.plot(year,profit, color='blue')
plt.title('Year against Sales and Profit')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(year,sales, color='red')
plt.title('Year against Sales')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.subplot(2,1,2)
plt.plot(year,profit, color='blue')
plt.title('Year against Profit')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.tight_layout()

```
