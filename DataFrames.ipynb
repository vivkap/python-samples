## DataFrames using Pandas

```python
import pandas as pd, numpy as np
```

```python
stu_dict = {'s_id': [1,2,4,5],
            's_name' : ['A','B','C','D'],
             'Region' : ['N','S','W','E']}
```

```python
sales_dict = {'stu_id': [1,2,4,6],
            'Course' : ['Py','R','Tab','DS'],
             'Sales' : [2520,1150,1850,2300]}
```

```python
stu_df = pd.DataFrame(stu_dict)
```

```python
sales_df = pd.DataFrame(sales_dict)
```

```python
stu_df.merge(sales_df,  how = "left")
```

```python
stu_df.merge(sales_df, left_on = 's_id',right_on = 'stu_id', how = "left")
```

```python
stu_df.merge(sales_df, left_on = 's_id',right_on = 'stu_id', how = "right")
```

```python
stu_df.merge(sales_df, left_on = 's_id',right_on = 'stu_id', how = "outer")
```
