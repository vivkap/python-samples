```python
df = pd.read_csv('HR_Data.csv', header=0)
```

```python
for row_index,row in df.iterrows():
    tags=row[0]
    field_value=row[8]
    json_body = [
        {
        "measurement": "HR_Data",
        "tags": {
            "Reference":tags
                },
        "fields": {
            "value" : field_value
               }
     }
    ]
    print(json_body)
    client.write_points(json_body)
    
```
